# Creating Sub plots using OOP

```python
fig,axes = plt.subplots()
axes.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=3,ncols=3)
plt.tight_layout() #fixing overlaps
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)
axes[0].plot(x,y)
axes[0].set_title('Firsrt Plot')

axes[1].plot(y,x)
axes[1].set_title('Seconf Plot')
```

```python
axes
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)
axes # list of subplots , we can iterate through
for i in axes:
    i.plot(x,y)
```

#  legends

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,label='X view')
ax.plot(y,x, label='Y view')

ax.set_title("Title")

ax.legend(loc=0) # location of legend on the map
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='blue')
ax.plot(x,y,color='green')
ax.plot(x,y,color='#FF8D01') # using hex code for potting the exact color
```
