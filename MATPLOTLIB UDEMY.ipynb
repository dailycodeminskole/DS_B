```python
plt.plot(x,y)
plt.plot(x,y, 'g-') ##  adding colors

# labelling the charts
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title(' Title Here')
```

# Class 2

```python
#  OO method
```

```python
fig = plt.figure() # creating an object fig , acting as blank canvas
axes  = fig.add_axes([0.1,0.1,0.8,0.8]) #L B W H arguments
axes.plot(x,y)
plt.show()
```

```python
plt.figure().add_axes([0.1,0.1,0.8,0.8]).plot(x,y)
plt.figure().add_axes([0.1,0.1,0.8,0.8]).set_xlabel("X Label")
plt.figure().add_axes([0.1,0.1,0.8,0.8]).set_ylabel("Y Label")
plt.figure().add_axes([0.1,0.1,0.8,0.8]).set_title("Title Here")
```

```python
# two sets of fig in one canvas
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])
axes1.plot(x,y)
axes2.plot(x,y)
axes2.set_title("Smaller Plaot")
axes1.set_title("Larger Plaot")
```
